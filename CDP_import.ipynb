```python
import pandas as pd

from flask import *
from __init__ import app

from flask.ext.sqlalchemy import SQLAlchemy
from apps.glyph.models import *

app.testing = True
client = app.test_client()
ctx = app.test_request_context()
ctx.push()
```

```python
# df = pd.DataFrame.from_csv("csvs/cdp_signs_modified.csv")
df = pd.read_excel('csvs/clean_CDP.xlsx', 'Sheet1', index_col='index', na_values=['NA'])
# replace NaN with None
signs = df.where((pd.notnull(df)), None)
```

```python
signs.head()
```

```python
# Add unique CDP Signs as Sign instances
uniq = list(signs['CDP SIGN REF NAME'].drop_duplicates().dropna())
new_signs = []
for each in uniq:
    s = Sign(sign_ref=each)
    new_signs.append(s)
db.session.add_all(new_signs)
db.session.commit()
```

```python
# Add unique Descriptions
descs = list(signs['Description'].drop_duplicates().dropna())
encoded = [s.encode('UTF8') for s in descs]
final = list(set(encoded))
new_descr = []
for each in final:
    s = Description(sign_ref=each)
    new_descr.append(s)
db.session.add_all(new_descr)
db.session.commit()
```

```python
# Add unique Oracc Descriptions
uniq = list(signs['ORACC sign'].drop_duplicates().dropna())
new_oracc = []
for each in uniq:
    s = Oracc(sign_ref=each.encode('utf8'))
    new_oracc.append(s)
db.session.add_all(new_oracc)
db.session.commit()
```

```python
# Add unique CDLI Archaic
uniq = list(set(list(signs['CDLI archaic'].drop_duplicates().dropna())))
new_cdli = []
for each in uniq:
    s = Cdli(sign_ref=each.encode('utf8'))
    new_cdli.append(s)
db.session.add_all(new_cdli)
db.session.commit()
```

```python
# Fix missing 'NA' values (record 3225-8 in import CSV)
# repeat for NA~a ~b ~c ~d
# TODO: add a "NA" entry for Sign, Oracc, and CDLI in the DB!
for s in ['NA~a', 'NA~b', 'NA~c', 'NA~d']:
    signs.ix[signs['CDLI archaic'] == s, 'CDP SIGN REF NAME'] = 'NA'
    signs.ix[signs['CDLI archaic'] == s, 'ORACC sign'] = 'NA'
    signs.ix[signs['CDLI archaic'] == s, 'Description'] = 'NA'
```

```python
to_add = list()

fields = [
    "variant_name",
    "form_description",
    "notes",
    "MesZL",
    "ELLes",
    "ZATU",
    "LAK",
    "UET_2",
    "ARM_XV",
    "Hinke",
    "Clay_BE_A_14",
    "Koenig_AfO_Bei_16",
    "Ranke_BE_A_61",
    "Schroeder_VS_12",
    "Clay_BE_A_10",
    "RSP",
    "Emar",
    "Schroder_VS_15",
    "HZL",
    "HA",
    "aBZL",
    "REC",
    "Labat",
    "KWU",
    "Fossey_pp",
]

for i, record in signs.iterrows():
    dict_record = dict(record)
    d = dict()
#     if dict_record.get('CDP SIGN REF NAME'):
    d['sign'] = Sign.query.filter_by(sign_ref=unicode(dict_record.get('CDP SIGN REF NAME')).encode('UTF8').strip()).one()
    if dict_record.get('Description'):
        d['description'] = Description.query.filter_by(sign_ref=unicode(dict_record.get('Description')).encode('UTF8').strip()).one()
    if dict_record.get('ORACC sign'):
        d['oracc'] = Oracc.query.filter_by(sign_ref=unicode(dict_record.get('ORACC sign')).encode('UTF8').strip()).one()
    if dict_record.get('CDLI archaic'):
        d['cdli'] = Cdli.query.filter_by(sign_ref=unicode(dict_record.get('CDLI archaic')).encode('UTF8').strip()).one()
    for f in fields:
        d[f] = dict_record.get(f)
    to_add.append(Cdp(**d))
```

```python
db.session.add_all(to_add)
db.session.commit()
```
